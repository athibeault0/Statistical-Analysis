```
#setting up the workspace
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
```

```
#downloading the dataset and saving it to "food"
!gdown 12Sd3fb832cvni2Mg5wJV0DcGj80e_bV4
food = pd.read_csv('/content/food.csv')
```

```
#the code below is to generate the statistics and statistical model for Pollution
food.describe()
```

```
#table with summary statistics of Pollution by Food_Category
from tabulate import tabulate
group = food.groupby('Food_Category')[['Pollution']].mean()
Pollution_Table = tabulate(tabular_data=group, headers='keys', tablefmt='fancy_grid')
print(Pollution_Table)
```

```
#importing other packages
import statsmodels.api as smapi
import statsmodels.formula.api as smf
import statsmodels.stats as stats
import matplotlib.pyplot as plt
#statistical model
model = smf.ols(formula='Pollution ~ Food_Category', data=food)
Fit_model = model.fit()
Fit_model.summary()
```

```
#Anderson Darling test for normality
stats.diagnostic.normal_ad(Fit_model.resid_pearson)
```

```
Fit_model.resid.hist()
```

```
smapi.qqplot(Fit_model.resid_pearson)
```

```
#Breusch-Pagan test for equal variance
stats.diagnostic.het_breuschpagan(Fit_model.resid_pearson, Fit_model.model.exog)
```

```
plt.scatter(Fit_model.fittedvalues, Fit_model.resid_pearson)
```

```
#since all the assumptions were violated, we use a HuberT estimate
import patsy
      
f = 'Pollution ~ Food_Category'
y, X = patsy.dmatrices(f, food, return_type='dataframe')
```

```
#robust linear model with the HuberT estimate
rlm_model = smapi.RLM(y, X, M=smapi.robust.norms.HuberT()).fit()
rlm_model.summary()
```
